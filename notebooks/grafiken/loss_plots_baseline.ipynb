```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Train Loss
data_train = pd.read_csv('../docs/baseline/train_epoch_logs_finetuned.csv')
epoch_progress_train = np.arange(1, len(data_train) + 1)

# Val Loss
data_val = pd.read_csv('../docs/baseline/val_epoch_logs_finetuned.csv')
epoch_progress_val = np.arange(1, len(data_val) + 1)


fig, axes = plt.subplots(1, 2, figsize=(16, 5), sharey=False)

# Plot: Train Loss
axes[0].plot(epoch_progress_train, data_train['train_loss'], color="#E69F00", linewidth=2)
axes[0].set_title('Train Loss während dem Training', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoche', fontsize=13)
axes[0].set_ylabel('Loss', fontsize=13)
axes[0].grid(True, color='lightgray', linestyle='--', linewidth=0.7, alpha=0.6)

axes[1].plot(epoch_progress_val, data_val['val_loss'], color="#E69F00", linewidth=2)
axes[1].set_title('Validation Loss während dem Training', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoche', fontsize=13)
axes[1].set_ylabel('Loss', fontsize=13)
axes[1].grid(True, color='lightgray', linestyle='--', linewidth=0.7, alpha=0.6)

plt.tight_layout()
plt.show()
```
